## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# total number churn and no churn customer
churn_df = df.groupby("Churn")["Churn"].count().reset_index(name="Total number")
print(f"Total number of customer is {df.Churn.count()}")
total_churn = df["Churn"].value_counts()[1]
print(f"Total number of churned customer is {total_churn}")
total_nochurn = df["Churn"].value_counts()[0]
print(f"Total number of not churned custmer is {total_nochurn}")
churn_df
````

````python
#data of churned customer
churn_full_df = df.loc[df.loc[:,"Churn"]==True]
#data of not churned customer
nochurn_full_df = df.loc[df.loc[:,"Churn"]==False]
````

````python
# % of total customers churning
perc_churn = df.loc[:,"Churn"].mean()*100
print(f"Percentage of customer churning : {round(perc_churn,2)}%")
````

````python
#statwise number of churn
churn_full_df = df.loc[df.loc[:,"Churn"]==True]
state_churn = churn_full_df.groupby("State")["Churn"].count().sort_values(ascending=False).reset_index(name="No.of churn")
print("Total number of churn of each state\n")
state_churn
````

````python
#Statewise maximum account length
state_max_acc_len = df.groupby("State")["Account length"].sum().sort_values(ascending=False).reset_index(name="Account length")
print("State maximum account length is",state_max_acc_len.loc[0,"State"])
print("State minimum account length is",state_max_acc_len.loc[50,"State"])
print("Maximum account length of each state\n")
state_max_acc_len
````

````python
#Total loss of each state due to churn
loss_state = churn_full_df.groupby("State").agg({"Total day charge":"sum","Total eve charge":"sum","Total night charge":"sum","Total intl charge":"sum"})
def add(a,b,c,d):
  return a+b+c+d
loss_state["Total charge"] = loss_state.apply(lambda x:add(x["Total day charge"],x["Total eve charge"],x["Total night charge"],x["Total intl charge"]),axis=1)
print("Total loss for company",loss_state["Total charge"].sum())
print("Total loss for company on each state\n")
loss_state['Total charge']
````

````python
#total number of international plan
print("Number of Customer with International plan active is",df.loc[:,"International plan"].value_counts()[1])
print("Number of Customer with International plan not active is",df.loc[:,"International plan"].value_counts()[0])
df.loc[:,"International plan"].value_counts()
````

````python
#percentage of churn in international plan
intl_plan = df.loc[df.loc[:,"International plan"]=="Yes"]
print("Number of customer with International plan active is",df.loc[:,"International plan"].value_counts()[1])
print("Number of customer churned with international plan active is",intl_plan["Churn"].value_counts()[1])
perc_churn_intl = (intl_plan["Churn"].value_counts()[1]/df.loc[:,"International plan"].value_counts()[1])*100
print("Percentage of customer churned in international plan",round(perc_churn_intl,2),"%")
````

````python
#total number of voice mail plan
print("Number of Customer with Voice mail plan active is",df.loc[:,"Voice mail plan"].value_counts()[1])
print("Number of Customer with Voice mail plan not active is",df.loc[:,"Voice mail plan"].value_counts()[0])
df.loc[:,"Voice mail plan"].value_counts()
````

````python
#percentage of churn in Voice mail plan
vml_plan = df.loc[df.loc[:,"Voice mail plan"]=="Yes"]
print("Number of customer with Voice mail plan active is",df.loc[:,"Voice mail plan"].value_counts()[1])
print("Number of customer churned with Voice mail plan active is",intl_plan["Churn"].value_counts()[1])
perc_churn_vml = (vml_plan["Churn"].value_counts()[1]/df.loc[:,"Voice mail plan"].value_counts()[1])*100
print("Percentage of customer churned in Voice mail plan",round(perc_churn_vml,2),"%")
````

````python
#number of customers in each area
df1 = df.groupby("Area code")["Area code"].count().reset_index(name="Total customer")
print("Number of customer in each area")
df1
````

````python
#number of customers churned in each area
df2 = churn_full_df.groupby("Area code")["Area code"].count().reset_index(name="Total customer")
print("Number of customer churned in each area")
df2
````

````python
#churn percentage for each area
perc_churn_408 = (churn_full_df.loc[:,"Area code"].value_counts()[408]/df.loc[:,"Area code"].value_counts()[408])*100
perc_churn_415 = (churn_full_df.loc[:,"Area code"].value_counts()[415]/df.loc[:,"Area code"].value_counts()[415])*100
perc_churn_510 = (churn_full_df.loc[:,"Area code"].value_counts()[510]/df.loc[:,"Area code"].value_counts()[510])*100
print("Percentage of customer churned in Area 408 is",round(perc_churn_408,2),"%")
print("Percentage of customer churned in Area 415 is",round(perc_churn_415,2),"%")
print("Percentage of customer churned in Area 510 is",round(perc_churn_510,2),"%")
````

````python
df.head(3)
````

````python
#charge per call
ch_day = (df["Total day charge"].sum()/df["Total day minutes"].sum())
ch_eve = (df["Total eve charge"].sum()/df["Total eve minutes"].sum())
ch_night = (df["Total night charge"].sum()/df["Total night minutes"].sum())
ch_intl = (df["Total intl charge"].sum()/df["Total intl minutes"].sum())

print("Charge per call at day time is",round(ch_day,2))
print("Charge per call at evening time is",round(ch_eve,2))
print("Charge per call at night time is",round(ch_night,2))
print("Charge per call for international call is",round(ch_intl,2))
````

````python
#groupby based on customer service calls
cs_call = df.groupby(["Customer service calls","Churn"])["Churn"].count()
print("This data shows churn data based on customer call")
cs_call
````

````python
# Combining of day,evening and night calls, mins and charges
df.loc[:,"Total calls"] = df.loc[:,["Total day calls","Total eve calls", "Total night calls"]].sum(axis=1)
df.loc[:,"Total minutes"] = df.loc[:,["Total day minutes","Total eve minutes", "Total night minutes"]].sum(axis=1)
df.loc[:,"Total charge"] = df.loc[:,["Total day charge","Total eve charge", "Total night charge"]].sum(axis=1)

df.loc[:,"Charge per minutes"] = df.loc[:,"Total charge"]/df.loc[:,"Total minutes"]
df.loc[:,"Charge per intl call"] = df.loc[:,"Total intl charge"]/df.loc[:,"Total intl minutes"]

df.head()
````
